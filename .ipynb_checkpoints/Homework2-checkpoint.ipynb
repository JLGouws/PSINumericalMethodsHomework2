```julia
Ylm(l, m, θ, φ) = sqrt((2 * l + 1) / (4 * π) * factorial(l - m) / factorial(l + m)) * Plm(cos(θ), l, m) * exp(im * m * φ) 
```

```julia
function flm(f, l, m)
    println([l, m])
    hcubature( x -> sin(x[1]) * f(x[1], x[2]) * conj(Ylm(l, m, x[1], x[2])), [0, 0], [π, 2 * π])
end
```

```julia
expandSphericalHarmonics(f, lₘₐₓ) = [[flm(f, l, m) for m in -l:l] for l in 0:lₘₐₓ]
```

```julia
ϕ₀SphericalCoeffs = expandSphericalHarmonics2(ϕ₀, 15)[1];
ψ₀SphericalCoeffs = expandSphericalHarmonics2(ψ₀, 15)[1];
```

```julia
function LaplacianSphericalCoefficients(coeffs, lₘₐₓ)
    for l in 0:lₘₐₓ
        for m in -l:l
            coeffs[(l, m)] *= - l * (l + 1)
        end
    end
    coeffs
end
```

```julia
function LaplacianSphericalCoefficientsArays(coeffs, lₘₐₓ)
    l = 0
    m = 0
    for a in 0:lₘₐₓ * (lₘₐₓ + 2)
        coeffs[a + 1] *= - l * (l + 1)
        if m == l
            l+=1
            m = -l
        else
            m+=1
        end
    end
    coeffs
end
```

```julia
function LaplacianSphericalCoefficientsArrays(coeffs)
    l = 0
    m = 0
    for a in 1:size(coeffs)[1]
        coeffs[a] *= - l * (l + 1)
        if m == l
            l+=1
            m = -l
        else
            m+=1
        end
    end
    coeffs
end
```

```julia
function flattenSHArray(shArray, lₘₐₓ)
    l = 0
    m = 0
    coeffVector::Array{ComplexF64} = []
    for a in 0:lₘₐₓ * (lₘₐₓ + 2)
        push!(coeffVector, shArray[(l, m)])
        if m == l
            l+=1
            m = -l
        else
            m+=1
        end
    end
    coeffVector
end
```

```julia
function flattenSHArray(shArray)
    l = 0
    m = 0
    coeffVector::Array{ComplexF64} = []
    for a in 1:size(shArray)[1]
        push!(coeffVector, shArray[(l, m)])
        if m == l
            l+=1
            m = -l
        else
            m+=1
        end
    end
    coeffVector
end
```

Here I solve the problem:
$$
\begin{align*}
    \partial_t \phi &= \psi\\
    \partial_t \psi &= \Delta \phi
\end{align*}
$$

```julia
chunk(arr, n) = [arr[i:min(i + n - 1, end)] for i in 1:n:length(arr)]

function waveSystem(U)
    SplitArray = chunk(U, Int(size(U)[1] // 2))
    ϕ = SplitArray[1]
    ψ = SplitArray[2]
    vcat(ψ, LaplacianSphericalCoefficientsArays(ϕ))
end
```

```julia
U0 = vcat( flattenSHArray(ϕ₀SphericalCoeffs, 15), flattenSHArray(ψ₀SphericalCoeffs, 15))
prob = ODEProblem((U,p,t) -> waveSystem(U), U0, (0.0, 10.0))
sol = solve(prob);
```

$$
\begin{align*}
    &\sum_{l = 0}^{l_{max}} (2 l + 1)\\
    = &2 \sum_{l = 0}^{l_{max}} l + \sum_{l = 0}^{l_{max}}1\\
    = & l (l + 1) + l
    =& l (l + 2)
\end{align*}
$$

```julia
function functionFromSphericalCoefficients(coeffs)
    l = 0
    m = 0
    lₘₐₓ = 0
    for a in 1:size(coeffs)[1]
        lₘₐₓ = l
        if m == l
            l+=1
            m = -l
        else
            m+=1
        end
    end
    function f(θ, ϕ)
        harmonics = computeYlm(θ, ϕ, lmax = lₘₐₓ)
        sum(harmonics .* coeffs)
    end
end
```
